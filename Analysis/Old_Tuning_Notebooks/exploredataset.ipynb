```python
# Reading in the data:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('../csvs/initial_orbital_elements.csv', index_col=0)
```

```python
df.tail()
```

```python
fig, ax = plt.subplots(figsize=(15,10))
cm = plt.cm.get_cmap('RdYlBu')
dfplot = df #df.loc[df["Norbits"] > 0.]
sc = ax.scatter(dfplot['beta12'], dfplot['beta23'], c=np.log10(dfplot['instability_time']), edgecolor='', vmin=8.9,s=10,cmap=cm)
ax.set_xlabel('beta12', fontsize=24)
ax.set_ylabel('beta23', fontsize=24)
plt.colorbar(sc)
```

```python
plt.hist(np.log10(df['instability_time']), 30)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
cm = plt.cm.get_cmap('RdYlBu')
dfplot = df #df.loc[df["Norbits"] > 0.]
sc = ax.scatter(dfplot['P2']/dfplot['P1'], dfplot['P3']/dfplot['P2'], c=np.log10(dfplot['instability_time']), edgecolor='', vmin=8.9,s=10,cmap=cm)
ax.set_xlabel('P2/P1', fontsize=24)
ax.set_ylabel('P3/P2', fontsize=24)
plt.colorbar(sc)
```
